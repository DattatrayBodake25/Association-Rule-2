```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori,association_rules

import warnings
warnings.filterwarnings('ignore')

%matplotlib inline
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/aishwaryamate/Datasets/main/Supermarket.csv')
```

```python
df
```

```python
df = df.drop('ID',axis=1)
```

```python
df
```

```python
def split(text):
    return text.split(',')
```

```python
a = 'Milk, bread, sauce'
```

```python
split(a)
```

```python
items =list( df['Products'].apply(split))
```

```python
items
```

```python
te = TransactionEncoder()
```

```python
endf = te.fit_transform(items)
```

```python
te.columns_
```

```python
endf
```

```python
data = pd.DataFrame(data=endf,columns=te.columns_)
```

```python
data
```

```python
data.replace(False,0,inplace=True)
data.replace(True,1,inplace=True)
```

```python
data
```

```python
ap = apriori(data,min_support=0.2,use_colnames=True)
```

```python
ap
```

```python
model = association_rules(ap,metric='confidence')
```

```python
model
```
